```python
results = pd.read_csv('/content/sample_data/results.csv', parse_dates=['date'])
```

```python
results.head(20)
```

```python
wc = results[results['tournament'] == 'FIFA World Cup'] #Selecionando somente as partidas de Copa
```

```python
wc.head(15)
```

filtrei somente as partidas de Copa do Mundo

```python
wc['year'] = results['date'].dt.year
```

```python
wc
```

As informações de time de casa e do time visitante não são importantes para essa análise que se prende ao saldo de gols por ano.

```python
home_scores = wc[['home_team','home_score','year']].rename({'home_team':'team','home_score':'score'}, axis='columns')
away_scores = wc[['away_team','away_score','year']].rename({'away_team':'team','away_score':'score'}, axis='columns')
```

```python
home_scores
```

```python
away_scores
```

Na visualização pretendida é pretendido saber quais os gols que o time fez por cada ano de copa, sem se importar se é dentro ou fora de casa.

```python
scores_wc = pd.concat([home_scores, away_scores]).reset_index(drop=True).fillna(0)
```

```python
scores_wc
```

```python
scores_wc.groupby(['team','year']).score.sum().unstack().fillna(0)
```

```python
acumulative_scores_wc = scores_wc.groupby(['team','year']).score.sum().unstack().fillna(0).cumsum(axis=1).reset_index()
```

```python
acumulative_scores_wc
```

```python
acumulative_scores_wc.shape
```

```python
acumulative_scores_wc[acumulative_scores_wc['team'] == 'United States']
```
